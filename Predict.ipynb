```python
import os
import lib.experimentize as E
from src.core import Core, Params

CODE = "PATTOM"

@E.experimentize(Params)
class Experiment(Core):
    def predict(self, *, method=E.param(as_func="feature_selection_{}"), dataset=E.param):
        train_x, train_y, test_x = self.get_data()
        selector, features = method(train_x, train_y)
        train_x, test_x = selector(train_x), selector(test_x)

        automl = self.auto_classifier()
        automl.fit(train_x, train_y)
        prediction = automl.predict_proba(test_x)[:, 1]
        
        os.makedirs("results", exist_ok=True)
        with open(f"results/{CODE}_{dataset}_prediction.txt", "w") as f:
            for p in prediction:
                f.write(f"{p}\n")
        
        with open(f"results/{CODE}_{dataset}_features.txt", "w") as f:
            for i in features:
                f.write(f"{i}\n")
```

```python
Experiment().predict(E.AsExperiment([Params(
    time_limit=180,
    dataset="artificial",
    method="Random Forest",
    max_features=15
)]))()
```
